```python
# Exercise 3

# Importing relevant libraries
import numpy as np
import matplotlib.pyplot as plt

# Make data
x = np.linspace(0,10,num=100)
y = np.log(x)
z = np.sin(x)

# Plot graphs of (y,x) and (z,x)
fig, ax = plt.subplots()
ax.plot(y, x)
plt.show()

fig, ax = plt.subplots()
ax.plot(z, x)
plt.show()
```

```python
# Exercise 4

# Make data
x = np.linspace(0,10,num=100)
y = np.linspace(0,10,num=100)
x, y = np.meshgrid(x, y)
z = np.log(x) + np.sin(y)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x, y, z)
plt.show()
```

```python
# Exercise 5

# Parameters values
alpha = 0.5
y = 10
p1 = 1
p2 = 2

# Define utility function
def utility_function(x1,alpha,y,p1,p2):
    x2 = (y-p1*x1)/p2 # Income not spent first good is spent on the second
    utility = x1**alpha * x2**(1-alpha)
    return utility

# Maximization to minimization
obj_func = lambda x1: -utility_function(x1,alpha,y,p1,p2)

from scipy.optimize import minimize_scalar
res = minimize_scalar(obj_func)
res.fun
res.x

x1_solved = res.x

def x2(x1_solved,alpha,y,p1,p2):
    x2 = (y-p1*x1_solved)/p2 # Income not spent first good is spent on the second
    return x2

x2_solved = x2(x1_solved,0.5,10,1,2)

print('The consumer would buy', x1_solved, 'units of x1 and', x2_solved, 'units of x2.')
```
